```python
# Dicionário de mapeamento para traduzir os nomes das colunas
traducao_colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Renomear as colunas do DataFrame
df.rename(columns=traducao_colunas, inplace=True)

# Exibir os nomes das colunas traduzidas
print(df.columns)
```

```python
# Mapeamento das categorias de senioridade para português
traducao_senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

# Substituir os valores na coluna 'senioridade'
df['senioridade'] = df['senioridade'].replace(traducao_senioridade)

# Exibir a contagem dos valores traduzidos
print(df['senioridade'].value_counts())
```

```python
# Mapeamento das categorias de contrato para português
traducao_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato Temporário',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}

# Substituir os valores na coluna 'contrato'
df['contrato'] = df['contrato'].replace(traducao_contrato)

# Exibir a contagem dos valores traduzidos
print(df['contrato'].value_counts())
```

```python
# Mapeamento das categorias de tamanho_empresa para português
traducao_tamanho_empresa = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}

# Substituir os valores na coluna 'tamanho_empresa'
df['tamanho_empresa'] = df['tamanho_empresa'].replace(traducao_tamanho_empresa)

# Exibir a contagem dos valores traduzidos
print(df['tamanho_empresa'].value_counts())
```

```python
# Mapeamento das categorias de remoto para português
traducao_remoto = {
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
}

# Substituir os valores na coluna 'remoto'
df['remoto'] = df['remoto'].replace(traducao_remoto)

# Exibir a contagem dos valores traduzidos
print(df['remoto'].value_counts())
```

Criar um dataframe do zero para exercitar criar uma média de valores faltantes de uma tabela.


```python

import numpy as np # importo a biblioteca numpy para manipualar valores

df_salarios = pd.DataFrame({
    'nome':['Ana','Bruno','Carlos','Daniele','Fulano',], # crio a coluna nome
    'salario':[4000,np.nan,5000,np.nan,100000]}) #Crio a coluna salario e deixo alguns valores nulos


df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
# logica da media acima: df_salario é a nossa base e salario_media é a nosso nova coluna
#então = dataframe_salarios = coluna salario(filtra valores nulus) e cria uma média dentro do df_salarios coluna salario com duas casas decimais de arredondamento



df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())


df_salarios# acresentei um salario de 100000 para entender como a mediana suavisa melhor em relação a média que pode não representar a realidade
```

```python
df_temperatura = pd.DataFrame({
    'dia':['Segunda','Terca','Quarta','Quinta','Sexta'],
    'temp':[30,np.nan,np.nan,28,27]
})

df_temperatura['preenchido_ffill'] = df_temperatura ['temp'].ffill()
'''o ffill completa com o valor anterior, no caso terça e quarta estavam nulos então
foi feito o complemento com o valor anterior
'''
df_temperatura
```
