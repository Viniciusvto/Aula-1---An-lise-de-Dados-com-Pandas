```python
# Dicionário de mapeamento para traduzir os nomes das colunas
traducao_colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Renomear as colunas do DataFrame
df.rename(columns=traducao_colunas, inplace=True)

# Exibir os nomes das colunas traduzidas
print(df.columns)
```

```python
df["senioridade"].value_counts() #analise de colunas (nivel_experiencia)
#Senior
#Pleno
#Junior
#executivo
```

```python
# Mapeamento das categorias de senioridade para português
traducao_senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

# Substituir os valores na coluna 'senioridade'
df['senioridade'] = df['senioridade'].replace(traducao_senioridade)

# Exibir a contagem dos valores traduzidos
print(df['senioridade'].value_counts())
```

```python
# Mapeamento das categorias de contrato para português
traducao_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato Temporário',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}

# Substituir os valores na coluna 'contrato'
df['contrato'] = df['contrato'].replace(traducao_contrato)

# Exibir a contagem dos valores traduzidos
print(df['contrato'].value_counts())
```

```python
# Mapeamento das categorias de tamanho_empresa para português
traducao_tamanho_empresa = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}

# Substituir os valores na coluna 'tamanho_empresa'
df['tamanho_empresa'] = df['tamanho_empresa'].replace(traducao_tamanho_empresa)

# Exibir a contagem dos valores traduzidos
print(df['tamanho_empresa'].value_counts())
```

```python
# Mapeamento das categorias de remoto para português
traducao_remoto = {
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
}

# Substituir os valores na coluna 'remoto'
df['remoto'] = df['remoto'].replace(traducao_remoto)

# Exibir a contagem dos valores traduzidos
print(df['remoto'].value_counts())
```
